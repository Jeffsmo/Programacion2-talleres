```python
import math  #libreria con operaciones matemáticas ya definidas...

# Definir una lista vacía para almacenar los datos
datos = []

# Función para mostrar el menú
def mostrar_menu():
    print("Menú:")
    print("1. Número de datos")
    print("2. Ingresar datos")
    print("3. Mostrar datos")
    print("4. Calcular promedio de datos")
    print("5. Calcular desviación estándar de los datos")
    print("6. Salir")

# Función para obtener el número de datos ingresados
def obtener_numero_datos():
    return len(datos)

# Función para ingresar datos a la lista
def ingresar_datos():
    dato = input("Ingrese un dato: ")
    # Validar que el dato ingresado sea un número decimal
    try:
        dato = float(dato)
    except ValueError:
        print("El valor ingresado no es válido.")
        return
    # Validar que el dato no esté duplicado en la lista
    if dato in datos:
        print("El dato ya se encuentra en la lista.")
        return
    datos.append(dato)
    print("El dato ha sido ingresado correctamente.")

# Función para mostrar los datos ingresados
def mostrar_datos():
    print("Datos ingresados:")
    for dato in datos:
        print(dato)

# Función para calcular el promedio de los datos
def calcular_promedio():
    if len(datos) == 0:
        print("No hay datos ingresados.")
        return
    suma = sum(datos)
    promedio = suma / len(datos)
    print("El promedio de los datos es:", promedio)

# Función para calcular la desviación estándar de los datos
def calcular_desviacion_estandar():
    if len(datos) == 0:
        print("No hay datos ingresados.")
        return
    suma = sum(datos)
    promedio = suma / len(datos)
    suma_cuadrados = sum([(dato - promedio) ** 2 for dato in datos])
    desviacion_estandar = math.sqrt(suma_cuadrados / len(datos))
    print("La desviación estándar de los datos es:", desviacion_estandar)

# Ciclo principal del programa
while True:
    mostrar_menu()
    opcion = input("Seleccione una opción: ")
    # Validar que la opción ingresada sea válida
    try:
        opcion = int(opcion)
    except ValueError:
        print("La opción ingresada no es válida.")
        continue
    if opcion < 1 or opcion > 6:
        print("La opción ingresada no es válida.")
        continue
    # Ejecutar la función correspondiente a la opción seleccionada
    if opcion == 1:
        print("Número de datos ingresados:", obtener_numero_datos())
    elif opcion == 2:
        ingresar_datos()
    elif opcion == 3:
        mostrar_datos()
    elif opcion == 4:
        calcular_promedio()
    elif opcion == 5:
        calcular_desviacion_estandar()
    elif opcion == 6:
        break

```
